```python
import numpy as np
import matplotlib.pyplot as plt

from qiskit import QuantumCircuit
from qiskit.circuit.library import RZGate, RXGate


n = 5 # No. of qubits
theta = 0.8   # rotation angle

# Start the circuit
qc = QuantumCircuit(n)

# Apply symmetric rotational X and rotational Z gates on each qubit
for qubit in range(n):
    qc.rx(theta, qubit)
    qc.rz(theta, qubit)

# Entangle qubits using CZ gates
for i in range(n - 1):
    qc.cz(i, i + 1) # entangle qubit i with qubit i+1


qc.draw(output = 'mpl')
```
